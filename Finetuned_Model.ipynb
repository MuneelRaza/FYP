```python
from transformers import pipeline
pipe = pipeline(task="text-generation", model=model_name, tokenizer=tokenizer, max_length=200)
```

```python
# Ignore warnings
import warnings
# Ignore the UserWarnings from transformers and torch
warnings.filterwarnings("ignore", category=UserWarning, module="transformers.generation.utils")
warnings.filterwarnings("ignore", category=UserWarning, module="torch.utils.checkpoint")

while True:

  # Run text generation pipeline with our next model
  prompt = input("Enter your query: ")

  if prompt == "Quit":
    break

  result = pipe(f"<s>[INST] {prompt} [/INST]")
  response = result[0]['generated_text']
  inst_end_index = response.find("[/INST]")

  # Extract the substring after [/INST]
  result_string = response[inst_end_index + len("[/INST]"):].strip()

  print(result_string)
```
